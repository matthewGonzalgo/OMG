## Define routine to make data array object from inputs

````python
def make_da_from_grd(grd_path, num_lat, num_lon, starting_lat, starting_lon, delta_lat, delta_lon):
    g = np.fromfile(grd_path, dtype = '<f4')
    g = np.reshape(g, (num_lat, num_lon))
    
    lat_tmp = np.arange(num_lat)*(delta_lat)
    lat = starting_lat + lat_tmp
    print(lat[0], lat[-1])
    print(len(lat))
    
    lon_tmp = np.arange(num_lon)*(delta_lon)
    lon = starting_lon + lon_tmp
    print(lon[0], lon[-1])
    print(len(lon))
    
    da = xr.DataArray(g, dims= ('lat','lon'), coords={'lat': lat, 'lon': lon})
    
    return da
````

## Interpolate height fields to two lines, AB and BC.  AB is from fjord to mountain.  BC is from fjord up a glacier

````python
## Define three points.  A in fjord, B on mountain, C up the glacier
A = ( 61.513292, -42.952035)
B = ( 61.499270, -42.920974)
C = ( 61.484357, -43.077507) 
````

````python
## Create arrays of lats and lons between AB and AC
ab_lat = np.linspace(A[0], B[0], 1000)
ab_lon = np.linspace(A[1], B[1], 1000)

ac_lat = np.linspace(A[0], C[0], 1000)
ac_lon = np.linspace(A[1], C[1], 1000)
````

````python
## Turn arrays into data array objects
da_ab_lat = xr.DataArray(ab_lat, dims='z')
da_ab_lon = xr.DataArray(ab_lon, dims='z')

da_ac_lat = xr.DataArray(ac_lat, dims='z')
da_ac_lon = xr.DataArray(ac_lon, dims='z')
````

````python
## interpolate height fields to AB and AC
da_18_ab = da_18.interp(lat=da_ab_lat, lon=da_ab_lon)
da_18_ab = da_18_ab.where(da_18_ab > -100, np.nan)

da_17_ab = da_17.interp(lat=da_ab_lat, lon=da_ab_lon)
da_17_ab = da_17_ab.where(da_17_ab > -100, np.nan)

#%%
da_18_ac = da_18.interp(lat=da_ac_lat, lon=da_ac_lon)
da_18_ac = da_18_ac.where(da_18_ac > -100, np.nan)

da_17_ac = da_17.interp(lat=da_ac_lat, lon=da_ac_lon)
da_17_ac = da_17_ac.where(da_17_ac > -100, np.nan)
````

````python
def get_utm_range(max_lat, max_lon, min_lat, min_lon):
    min_x = utm.from_latlon(min_lat, min_lon)[0]
    max_x = utm.from_latlon(max_lat, max_lon)[0]
    dx = abs(max_x - min_x) # Shouldn't have to be absolute value

    min_y = utm.from_latlon(min_lat, min_lon)[1]
    max_y = utm.from_latlon(max_lat, max_lon)[1]
    dy = abs(max_y - min_y) # Shouldn't have to be absolute value

    return (min_x, max_x, dx, min_y, max_y, dy)
````
